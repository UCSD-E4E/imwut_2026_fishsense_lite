## Calibration
Below we go through calibration

```python
def point_to_plane(point: np.ndarray, std: float = 0.1) -> np.ndarray:
    random_normal = np.random.normal(0, std, (3,))
    random_normal /= np.linalg.norm(random_normal)
    
    if random_normal[2] < 0:
        random_normal = -random_normal

    return point, random_normal
```

```python
normals = []

for point in p.T:
    _, normal = point_to_plane(point)

    normals.append(normal)

normals = np.array(normals)
normals
```

```python
def atanasov_calibration_method(ps: np.ndarray):
    """
    Nikolay's method for laser calibration.
    Inputs:
     - ps: the laser points
    Output: the 5-vector of the laser parameters, with the first 3 being the orientation,
            and the final two being the x and y coordinates of the laser origin
    """
    avg_alpha = np.zeros((3,))
    params = np.zeros((5,))
    for i in range(ps.shape[0]):
        for j in range(ps.shape[0]):
            if i != j:
                v = ps[i] - ps[j]
                if v[2] < 0:
                    v = -v
                avg_alpha += v

    avg_alpha /= np.linalg.norm(avg_alpha)
    if avg_alpha[2] < 0:
        avg_alpha = -avg_alpha

    centroid = np.mean(ps, axis=0)
    print(centroid)
    scale_factor = centroid[2] / avg_alpha[2]
    params[0:3] = avg_alpha
    params[3:5] = centroid[0:2] - scale_factor * avg_alpha[0:2]
    return params
```

```python
def f(B, x):
    '''Linear function y = m*x + b'''
    # B is a vector of the parameters.
    # x is an array of the current x values.
    # x is in the same format as the x passed to Data or RealData.
    #
    # Return an array in the same format as y passed to Data or RealData.
    return B[0]*x + B[1]
```

```python
np.random.seed(0)

calibrations_by_stds = []
stds = [0.0, 0.5, 1.0, 2.0, 5.0, 10.0]

fig, axs = plt.subplots(3, 2)
fig.suptitle('Noisy Points Visualization')

for idx, std in enumerate(tqdm(stds)):
    x, y = divmod(idx, 2)
    ax = axs[x, y]

    s_noisy = np.random.normal(s, std, s.shape)
    rays = inverted_camera_intrinsics @ s_noisy
    scale_factor = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays)
    p_noisy = rays * scale_factor

    ax.scatter(s_noisy[0, :], s_noisy[1, :], c='blue', s=1, label='Noisy Points')

    model = odr.Model(f)
    data = odr.Data(s_noisy[0, :], s_noisy[1, :])
    odr_run = odr.ODR(data, model, beta0=[1.0, 1.0])
    results = odr_run.run()

    s_reconstructed = np.zeros_like(s_noisy)
    s_reconstructed[0, :] = results.xplus
    s_reconstructed[1, :] = results.y
    s_reconstructed[2, :] = 1
    rays_reconstructed = inverted_camera_intrinsics @ s_reconstructed
    scale_factor_reconstructed = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays_reconstructed)
    p_reconstructed = rays_reconstructed * scale_factor_reconstructed

    if np.allclose(p_noisy, p_reconstructed):
        print("Reconstruction matches noisy points exactly.")

    # s_pixel_noisy = np.random.normal(s_pixel, std, s_pixel.shape)
    # s_pixel_noisy = np.round(s_pixel_noisy)
    # rays_pixel = inverted_camera_intrinsics @ s_pixel_noisy
    # scale_factor_pixel = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays_pixel)
    # p_pixel_noisy = rays_pixel * scale_factor_pixel

    calibrations = []
    calibrations_reconstructed = []
    calibrations_pixels = []
    for n in tqdm(list(range(2, STEP_COUNT))):
        idxs = np.random.choice(np.arange(STEP_COUNT), n, replace=False)

        local_laser_origin, local_laser_orientation = calibrate_laser(p_noisy[:, idxs].T)
        calibration_noisy = np.concatenate([local_laser_orientation, local_laser_origin[:2]])

        local_laser_origin_reconstructed, local_laser_orientation_reconstructed = calibrate_laser(p_reconstructed[:, idxs].T)
        calibration_reconstructed = np.concatenate([local_laser_orientation_reconstructed, local_laser_origin_reconstructed[:2]])

        if std != 0 and np.allclose(calibration_noisy, calibration_reconstructed):
            print("Match:", calibration_noisy, calibration_reconstructed)

        # local_laser_origin_pixel, local_laser_orientation_pixel = calibrate_laser(p_pixel_noisy[:, idxs].T)
        # calibration_pixel_noisy = np.concatenate([local_laser_orientation_pixel, local_laser_origin_pixel[:2]])

        calibrations.append(calibration_noisy)
        calibrations_reconstructed.append(calibration_reconstructed)
        # calibrations_pixels.append(calibration_pixel_noisy)

    ax.set_title(f'Noisy Points (Std {std})')
    ax.set_xlabel('Pixel X')
    ax.set_ylabel('Pixel Y')

    ax.scatter(s[0, :], s[1, :], c='purple', s=1, label='Original Points')
    ax.scatter(results.xplus, results.y, c='red', s=1, label='ODR Fit')

    calibrations_by_stds.append((std, calibrations, calibrations_reconstructed, calibrations_pixels))
```

```python
for std, arrays, _ in calibrations_by_stds:
    percent_error = (np.array(arrays) - np.tile(laser_params, (len(arrays), 1))) / np.tile(laser_params, (len(arrays), 1))

    direction_percent_error = percent_error[:, ][:, 0:3]
    position_percent_error = np.mean(percent_error[:, ][:, 3:5], axis=1)

    plt.plot(np.arange(2, STEP_COUNT), position_percent_error, label=f'Std {std} Position Percent Error')

plt.title('Position Percent Error vs Number of Points Used for Calibration (Continuous)')
plt.xlabel('Number of Points Used for Calibration')
plt.ylabel('Percent Error')
plt.legend()
```

```python
for std, arrays, arrays_reconstructed, _ in calibrations_by_stds:
    mean_errors_noisy = []
    mean_errors_reconstructed = []

    for calibration_noisy, calibration_reconstructed in zip(arrays, arrays_reconstructed):
        laser_axis_noisy = calibration_noisy[0:3]
        
        laser_origin_noisy = np.zeros((3,))
        laser_origin_noisy[0:2] = calibration_noisy[3:5]
        laser_origin_noisy[2] = 0.0

        laser_axis_reconstructed = calibration_reconstructed[0:3]

        laser_origin_reconstructed = np.zeros((3,))
        laser_origin_reconstructed[0:2] = calibration_reconstructed[3:5]
        laser_origin_reconstructed[2] = 0.0

        projected_points = inverted_camera_intrinsics @ s
        norms = np.linalg.norm(projected_points, axis=0)
        final_laser_axis = -projected_points / norms

        point_constants_noisy = (
				(final_laser_axis.T @ laser_origin_noisy)
			 	- (laser_axis_noisy.T @ laser_origin_noisy) * (laser_axis_noisy.T @ final_laser_axis)
			) / (1 - (laser_axis_noisy.T @ final_laser_axis) ** 2)
        world_points_noisy = np.tile(point_constants_noisy, (3, 1)) * final_laser_axis
        mean_error_noisy = np.mean(np.sqrt(np.sum((world_points_noisy - p) ** 2, axis=1)))

        point_constants_reconstructed = (
                (final_laser_axis.T @ laser_origin_reconstructed)
                - (laser_axis_reconstructed.T @ laser_origin_reconstructed) * (laser_axis_reconstructed.T @ final_laser_axis)
            ) / (1 - (laser_axis_reconstructed.T @ final_laser_axis) ** 2)
        world_points_reconstructed = np.tile(point_constants_reconstructed, (3, 1)) * final_laser_axis
        mean_error_reconstructed = np.mean(np.sqrt(np.sum((world_points_reconstructed - p) ** 2, axis=1)))
        
        mean_errors_noisy.append(mean_error_noisy)
        mean_errors_reconstructed.append(mean_error_reconstructed)

    plt.plot(np.arange(2, STEP_COUNT), mean_errors_noisy, label=f'Std {std} Mean Reconstruction Error (Noisy)')
    plt.plot(np.arange(2, STEP_COUNT), mean_errors_reconstructed, label=f'Std {std} Mean Reconstruction Error (Reconstructed)')

plt.title('Mean Reconstruction Error vs Number of Points Used for Calibration (Continuous)')
plt.xlabel('Number of Points Used for Calibration')
plt.ylabel('Mean Reconstruction Error (m)')
plt.legend()
```
