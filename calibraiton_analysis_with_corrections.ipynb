```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm.notebook import tqdm
from fishsense_core.laser import calibrate_laser
from scipy import odr
from camera import calculate_normals
from plots import plot_position_percent_error, plot_mean_reconstruction_error, plot_mean_z_percent_error
```

## Setup of Simulation
The following reasonable defaults that match the systems we tested against.

```python
IMAGE_WIDTH, IMAGE_HEIGHT = 4014, 3016
FOCAL_LENGTH_PX = 2850

IMAGE_WIDTH, IMAGE_HEIGHT, FOCAL_LENGTH_PX
```

```python
camera_intrinsics = np.array([
    [FOCAL_LENGTH_PX, 0, IMAGE_WIDTH / 2],
    [0, FOCAL_LENGTH_PX, IMAGE_HEIGHT / 2],
    [0, 0, 1]
])
inverted_camera_intrinsics = np.linalg.inv(camera_intrinsics)

camera_intrinsics, inverted_camera_intrinsics
```

```python
laser_position = np.array([-0.04, -0.11, 0])
laser_direction = np.array([1e-10, 1e-10, 1])

laser_params = np.concatenate([laser_direction, laser_position[:2]])

laser_position, laser_direction, laser_params
```

## Next we setup data we can use for testing

```python
STEP_COUNT = 1000
t = np.linspace(0.5, 30, STEP_COUNT)

STEP_COUNT, t
```

```python
p = laser_position[:, np.newaxis] + t[np.newaxis, :] * laser_direction[:, np.newaxis]

p
```

```python
s = p / p[2, :]
s = camera_intrinsics @ s

s
```

```python
s_pixel = np.round(s)

s_pixel
```

Since our laser produces a line, the projection of the laser line is itself a line in the pinhole camera model.

```python
plt.scatter(s[0, :], s[1, :], c=p[2, :], cmap='viridis', marker='o')
plt.scatter(s_pixel[0, :], s_pixel[1, :], c=p[2, :], cmap='plasma', marker='o')
```

## Calibration
Below we go through calibration

```python
normals = calculate_normals(p)

normals
```

```python
def f(B, x):
    '''Linear function y = m*x + b'''
    # B is a vector of the parameters.
    # x is an array of the current x values.
    # x is in the same format as the x passed to Data or RealData.
    #
    # Return an array in the same format as y passed to Data or RealData.
    return B[0]*x + B[1]
```

```python
np.random.seed(0)

calibrations_by_stds = []
stds = [0.0, 0.5, 1.0, 2.0, 5.0, 10.0]

fig, axs = plt.subplots(3, 2)
fig.suptitle('Noisy Points Visualization')

for idx, std in enumerate(tqdm(stds)):
    x, y = divmod(idx, 2)
    ax = axs[x, y]

    s_noisy = np.random.normal(s, std, s.shape)
    rays = inverted_camera_intrinsics @ s_noisy
    scale_factor = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays)
    p_noisy = rays * scale_factor

    ax.scatter(s_noisy[0, :], s_noisy[1, :], c='blue', s=1, label='Noisy Points')

    model = odr.Model(f)
    data = odr.Data(s_noisy[0, :], s_noisy[1, :])
    odr_run = odr.ODR(data, model, beta0=[1.0, 1.0])
    results = odr_run.run()

    s_reconstructed = np.zeros_like(s_noisy)
    s_reconstructed[0, :] = results.xplus
    s_reconstructed[1, :] = results.y
    s_reconstructed[2, :] = 1
    rays_reconstructed = inverted_camera_intrinsics @ s_reconstructed
    scale_factor_reconstructed = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays_reconstructed)
    p_reconstructed = rays_reconstructed * scale_factor_reconstructed

    s_pixel_noisy = np.random.normal(s_pixel, std, s_pixel.shape)
    s_pixel_noisy = np.round(s_pixel_noisy)
    rays_pixel = inverted_camera_intrinsics @ s_pixel_noisy
    scale_factor_pixel = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays_pixel)
    p_pixel_noisy = rays_pixel * scale_factor_pixel

    model_pixel = odr.Model(f)
    data_pixel = odr.Data(s_pixel_noisy[0, :], s_pixel_noisy[1, :])
    odr_run_pixel = odr.ODR(data_pixel, model_pixel, beta0=[1.0, 1.0])
    results_pixel = odr_run_pixel.run()

    s_pixel_reconstructed = np.zeros_like(s_pixel_noisy)
    s_pixel_reconstructed[0, :] = results_pixel.xplus
    s_pixel_reconstructed[1, :] = results_pixel.y
    s_pixel_reconstructed[2, :] = 1
    rays_pixel_reconstructed = inverted_camera_intrinsics @ s_pixel_reconstructed
    scale_pixel_factor_reconstructed = np.einsum('ij,ij->j', normals.T, p) / np.einsum('ij,ij->j', normals.T, rays_pixel_reconstructed)
    p_pixel_reconstructed = rays_pixel_reconstructed * scale_pixel_factor_reconstructed

    calibrations = []
    calibrations_reconstructed = []
    calibrations_pixels = []
    calibrations_reconstructed_pixels = []
    for n in tqdm(list(range(2, STEP_COUNT))):
        idxs = np.random.choice(np.arange(STEP_COUNT), n, replace=False)

        local_laser_origin, local_laser_orientation = calibrate_laser(p_noisy[:, idxs].T)
        calibration_noisy = np.concatenate([local_laser_orientation, local_laser_origin[:2]])

        local_laser_origin_reconstructed, local_laser_orientation_reconstructed = calibrate_laser(p_reconstructed[:, idxs].T)
        calibration_reconstructed = np.concatenate([local_laser_orientation_reconstructed, local_laser_origin_reconstructed[:2]])

        local_laser_origin_pixel, local_laser_orientation_pixel = calibrate_laser(p_pixel_noisy[:, idxs].T)
        calibration_pixel_noisy = np.concatenate([local_laser_orientation_pixel, local_laser_origin_pixel[:2]])

        local_laser_origin_reconstructed_pixel, local_laser_orientation_reconstructed_pixel = calibrate_laser(p_pixel_reconstructed[:, idxs].T)
        calibration_reconstructed_pixel = np.concatenate([local_laser_orientation_reconstructed_pixel, local_laser_origin_reconstructed_pixel[:2]])

        calibrations.append(calibration_noisy)
        calibrations_reconstructed.append(calibration_reconstructed)
        calibrations_pixels.append(calibration_pixel_noisy)
        calibrations_reconstructed_pixels.append(calibration_reconstructed_pixel)

    ax.set_title(f'Noisy Points (Std {std})')
    ax.set_xlabel('Pixel X')
    ax.set_ylabel('Pixel Y')

    ax.scatter(results.xplus, results.y, c='red', s=1, label='ODR Fit')

    calibrations_by_stds.append((std, calibrations, calibrations_reconstructed, calibrations_pixels, calibrations_reconstructed_pixels))
```
